```python
#필요한 라이브러리 불러오기
import numpy as np
import pandas as pd
import random
import copy
```

```python
df = pd.read_csv("data/iris.csv") #데이터셋 불러오기
del df['variety'] #불필요한 열 제거
a=list(df.index) #행의 인덱스를 리스트자료형으로 나타내기
first=[0,0,0] #초기화(3개의 centroid)
centroid = {} 
for i in range(3): 
    centroid[i] = [0,0,0,0] #초기화(attribute 4개)     
```

```python
#중복되지 않은 랜덤한 centorid에 attribute값 대입
def data_centroid(first,centorid,df): 
    while 1:
        for i in range(3):
            first[i]=random.choice(a) #행의 인덱스들 중 1개를 대입
        if first[0]!=first[1]!=first[2]: # 중복되지 않게
            break
    for j in range(3):
        centorid[j]=df.iloc[first[j],:] #행의 인덱스값에 대응되는 attribute값 대입
    return first,centroid
```

```python
#각 centroid와의 거리를 나타내는 열 추가
for i in range(3):
    df['distance_from_centroid{}'.format(i)]=0
```

```python
#거리를 구하는 함수 생성
def distance(df,centroid):
    for i in range(3): #만약 i=0이면 첫번째 centroid의 대응되는 attribute와 Data들과의 거리를 구할 수 있다.
        df['distance_from_centroid{}'.format(i)]=( 
            np.sqrt(
            (df.iloc[:,0]-centroid[i][0])**2
            +(df.iloc[:,1]-centroid[i][1])**2
            +(df.iloc[:,2]-centroid[i][2])**2
            +(df.iloc[:,3]-centroid[i][3])**2
            )
         )
    return df
```

```python
distance(df,centroid)
```

```python
df['cluster'] = 0 #어느 Cluster에 할당되었는지를 보여주기 위한 열 추가

#데이터들을 할당하는 함수 생성
def assignment(df):
    #세가지 값의 거리를 하나의 리스트에 대입
    distance_from_all = ['distance_from_centroid{}'.format(i) for i in range(3)]
    #그 값중 가장작은 값을 대입
    df['cluster'] = df.loc[:,distance_from_all].idxmin(axis=1) 
    #숫자부분만 남기기위해 lstrip으로 제거
    df['cluster'] = df['cluster'].map(lambda x: int(x.lstrip('distnace_from'))) 
    return df
```

```python
assignment(df)
```

```python
#평가값을 구하는 함수 생성
def Value_cal(df):
    distance_min=[]
    for i in range(150):
        distance_min.append(min(list(df.iloc[i,4:7]))) #거리중 가장 작은 값을 리스트에 추가
    a=sum(distance_min) #리스트들을 모두 더한다
    return a
```

```python
Value_1=Value_cal(df)
```

```python
#중심데이터를 바꿔주는 함수 생성
def centroid_update(first,df):
    while 1:
        random_pick = random.randint(0,2) #3가지 데이터들중 1개를 뽑기위한 변수 생성
        random_pick_data_cluster = df.loc[first[random_pick],'cluster']
        original_data = copy.deepcopy(first) #deep copy를 통해 first가 바뀌여도 original_data는 바뀌지 않음
        first[random_pick] = random.choice(a) #뽑힌 초기중심 값을 다른 값으로 랜덤하게 바꿈
        if df.loc[first[random_pick],'cluster']==random_pick_data_cluster and original_data!=first: 
            break #바뀌기 전 Cluster와 바뀌 후 Cluster가 같은지 비교 & 인덱스 비교
    return first
```

```python
centroid_update(first,df)
```

```python
#위 과정을 반복하면서 최적의 평가값을 구해냄
Value_list=[]
k=0
while k<500: #1000번 반복
    k=k+1
    first=centroid_update(first,df)
    for i in range(3):
        centroid[i] = df.iloc[first[i],:]
    df=distance(df,centroid)
    df=assignment(df)
    Value=Value_cal(df)
    #list 함수중 append를 이용해 리스트 값을 추가
    Value_list.append(Value) 
min(Value_list)
```

```python
#silhouette 평가 값 산출과정
for i in range(150): #i번재 데이터를 기준으로
    for j in range(150): #모든 데이터들과 거리를 구한다.
            z=np.sqrt(
            (df.iloc[i,0]-df.iloc[j,0])**2
            +(df.iloc[i,1]-df.iloc[j,1])**2
            +(df.iloc[i,2]-df.iloc[j,2])**2
            +(df.iloc[i,3]-df.iloc[j,3])**2)
            df.loc[j,'distance_from_data{}'.format(i)]=z
```

```python
#각 데이터들마다 자신이 속한 Cluster의 거리를 x리스트에 나타냄
x=[]
for i in range(150): #i는 몇 번째 데이터인지
    a=df.loc[i,'cluster']
    dis=[]
    for j in range(150): #모든 데이터들에 대해
            if df.loc[j,'cluster']==a: #같은 cluster인 경우.
                dis.append(df.iloc[j,8+i]) #dis리스트에 거리 값 추가(8번째 열이 distance_from_data0)
    x.append(sum(dis)) #거리 값들을 더한 값을 x리스트에 추가
```

```python
#각 데이터들마다 자신이 속하지 않은 cluster의 거리를 y리스트에 나타냄
y=[]
for i in range(150): #i는 몇 번째 데이터인지
    other_cluster={0:0,1:0,2:0} #복수개의 cluster와의 거리를 나타내기위해 딕셔너리자료형{key:Value} 선언
    a=df.loc[i,'cluster']
    dis=[]
    for j in range(150):
            if df.loc[j,'cluster']!=a: #속하지않은 cluster의 경우
                #딕셔너리[key]=Value 형태를 사용. 반복적으로 그 값을 더해나간다.
                other_cluster[df.loc[j,'cluster']]=other_cluster[df.loc[j,'cluster']]+df.iloc[j,8+i]
    #3개의 cluster를 딕셔너리로 선언했기에 만약 자신의 클러스터일경우를 대비 (min값에 영향)
    if other_cluster[0]== 0:
        other_cluster[0]=999999
    elif other_cluster[1]==0:
        other_cluster[1]=999999
    elif other_cluster[2]==0:
        other_cluster[2]=999999        
    y.append(min(other_cluster[0],other_cluster[1],other_cluster[2]))                    
```

```python
silhouette_list=[]
for i in range(150):
    sil=(y[i]-x[i])/max(x[i],y[i])
    silhouette_list.append(sil)
sum(silhouette_list)/150
```
